## Overview
In this project, our goal is to build a binary classification model that predicts whether an arrest will occur after a Terry Stop. A Terry Stop refers to a brief detention by law enforcement officers based on reasonable suspicion, even in the absence of clear evidence for full-blown arrests. Our dataset includes information about the presence of weapons, time of day, and other relevant features.

## 1:Business Understanding
### Problem Statement
Law enforcement agencies require insights into the factors influencing arrest outcomes during Terry Stops. By predicting whether an arrest will occur, we can enhance decision-making and potentially reduce unnecessary detentions.

### Specific Objectives
1. Develop a robust binary classification model.
2. Identify key features that contribute to arrest decisions.
3. Investigate the role of gender and race in arrest outcomes.

## Research Questions
1. What factors are most influential in determining whether an arrest occurs after a Terry Stop?
2. How does the presence of weapons impact arrest outcomes?
3. Is there evidence of bias related to gender or race?

## Success Criteria
1. Achieve an accuracy of at least 80% in predicting arrest outcomes.
2. Provide actionable insights for law enforcement agencies.

## 2:Data Understanding
### Data Description
Our dataset includes the following features:
- **Arrest Outcome:** Binary (1 for arrest made, 0 for no arrest).
- **Presence of Weapons:** Indicates whether weapons were found during the stop.
- **Time of Day:** Categorized into morning, afternoon, evening, and night.
- **Other Relevant Features:** Additional contextual information.

```python
data.head()
```

After further consideration and analysis'potential_drops' would likely make the model more complex without necessarily improving its performance due several reasons:

 - Overfitting and Complexity:

Larger Models: More features require a larger model, increasing the risk of overfitting. This means the model performs well on the training data but poorly on unseen data.
Training Time: Training a complex model with many features takes longer and requires more computational resources.

- Redundant Information:

Overlapping Features: Some features might capture similar information. For example, "Reported Date" and "Reported Time" can be combined into "Day of Week" or "Time of Day."
Difficult Categorical Features: Categorical features with many categories can be challenging for models to learn from effectively.

- Privacy Concerns:

Location Information: Features like "Officer Squad," "Precinct," etc. might contain sensitive location information. Aggregating them to larger geographical units might be necessary.


Before handling missing values seeing a correlation between the columns'Stop Resolution' and 'Arrest Flag' we can create a binary 'Arrested' column

```python
#Condition for arrest based on Stop Resolution and Arrest Flag
arrested_condition = (data['Stop Resolution'] == 'Arrest') | (data['Arrest Flag'] == 'Y')

# Create the new 'Arrested' column with 1 for arrest and 0 for no arrest
data['Arrested'] = np.where(arrested_condition, 1, 0)

data.drop('Subject Age Group', axis=1, inplace=True)

data.head()
```

### 3.1.2 Handling missing values

```python

print(data.isnull().sum())  # Count missing values in each column
```

From this we conclude that Weapon Type has a high number of missing values (32565).From the source :Type of weapon, if any, identified during a search or frisk of the subject. Indicates "None" if no weapons was found thus missing data signifies no frisk/weapon

```python
data['Frisk Flag'] = data['Frisk Flag'].map(lambda x: 1 if x == 'Y' else 0)
data['Weapon Type'] = data['Weapon Type'].map(lambda x: 'None' if x == '-' else x)
```

```python
print(data.isnull().sum())  # Count missing values in each column
```

For Model Compatability reasons we convert and time to numerical "month" and "hour" columns 

```python
data['Reported Month'] = data['Reported Date'].map(lambda x: int(x[5:7]))
data.drop('Reported Date', axis=1, inplace=True)

data['Reported Hour'] = data['Reported Time'].map(lambda x: int(x[:2]))
data.drop('Reported Time', axis = 1, inplace=True)

data.head()
```

## 4:Modeling 
 

So as the models to treat all features equally we shall sclae the range of independant variables or features of the data

Below SMOTE is used to address class imbalance by oversampling the minority class ("Arrested" in this case)

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.impute import SimpleImputer
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
import numpy as np

# Separate features and target
X = data.drop("Arrested", axis=1)
y = data["Arrested"]

# Handle categorical features
categorical_cols = ["Stop Resolution", "Weapon Type", "Officer Gender", "Officer Race",
                    "Subject Perceived Race", "Subject Perceived Gender", "Age Group"]

# One-hot encode categorical features (recommended for most models)
encoder = OneHotEncoder(handle_unknown='ignore')  
X_encoded = pd.DataFrame(encoder.fit_transform(X[categorical_cols]))

# Ensure feature names are all strings (if applicable)
if not all(isinstance(col, str) for col in X.columns):
    X.columns = X.columns.astype(str)  # Convert to strings if necessary

# Handle potential string values in numerical columns 
for col in X.columns:
    if X[col].dtype == object:  
        try:
           
            X[col] = pd.to_numeric(X[col], errors='coerce')  
        except:
            pass  
# Handle missing values 
if X.isnull().sum().any():
    imputer = SimpleImputer(strategy="most_frequent")  
    X = imputer.fit_transform(X)

# Oversample the minority class using SMOTE
smote = SMOTE()
X_resampled, y_resampled = smote.fit_resample(X, y)

# Split the resampled data into training and testing sets (stratified split)
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled,
                                                    test_size=0.2, stratify=y_resampled, random_state=42)

# Print shapes of training and testing sets for confirmation
print(f"X_train shape: {X_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"y_test shape: {y_test.shape}")
```

### Baseline Models

- Majority Class Classifier: This predicts the majority class for all instances, providing a baseline performance to compare with actual models.
- Random Classifier: This predicts labels randomly, another baseline for comparison.

```python
from sklearn.dummy import DummyClassifier

# Majority class classifier (predicts majority class for all instances)
majority_clf = DummyClassifier(strategy="most_frequent")
majority_clf.fit(X_train, y_train)
y_pred_majority = majority_clf.predict(X_test)

# Random classifier (predicts labels randomly)
random_clf = DummyClassifier(strategy="uniform")
random_clf.fit(X_train, y_train)
y_pred_random = random_clf.predict(X_test)


from sklearn.metrics import classification_report

print("Majority Class Classifier Report:")
print(classification_report(y_test, y_pred_majority))

print("\nRandom Classifier Report:")
print(classification_report(y_test, y_pred_random))
```

### Model Tuning
- DescisionTree

Here we search through a defined grid of hyperparameter values (param_grid) to find the combination that yields the best performance on a validation set (specified by cv=5). This helps fine-tune the decision tree model for better accuracy.

```python
from sklearn.model_selection import GridSearchCV


param_grid = {
    'max_depth': [3, 5, 7],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 3, 5]
}
```

```python
# Create a decision tree classifier
clf = DecisionTreeClassifier()

# Use GridSearchCV to find the best hyperparameters
grid_search = GridSearchCV(estimator=clf, param_grid=param_grid, cv=5)
grid_search.fit(X_train, y_train)
```

```python
# Get the best model and score
best_model = grid_search.best_estimator_
best_score = grid_search.best_score_

# Use the best model for prediction
y_pred = best_model.predict(X_test)

from sklearn.metrics import accuracy_score
print('Accuracy score:', accuracy_score(y_test, y_pred))
```
